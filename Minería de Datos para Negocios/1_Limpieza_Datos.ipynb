##### 5 Agrear columna, operaciones basicas entre columnas

```python
#Agregar una columna igual a 1 si la columna gender es igual a 'M' y 0 si es 'F'
datos2['Genero'] = datos['Gender'].apply(lambda x: 1 if x == 'M' else 0)

# Agregar una columna con el promedio de las columnas numéricas
#datos2['Promedio'] = datos[numeric_columns].mean(axis=1)

# Asegurarse de que todas las columnas involucradas en operaciones aritméticas sean numéricas
numeric_columns = datos.select_dtypes(include=['number']).columns
datos2['Promedio'] = datos[numeric_columns].mean(axis=1)

# Agregar una columna con la suma de las columnas numéricas
datos2['Suma'] = datos[numeric_columns].sum(axis=1)

#Agrear una columa con la suma de dos columnas
datos2['Suma2'] = datos['Customer_Age'] + datos['Dependent_count']

# Agregar una columna con la resta de las columnas
datos2['Resta'] = datos['Customer_Age'] - datos['Dependent_count']

#Agregar una columna con la multiplicación de las columnas
datos2['Multiplicacion'] = datos['Customer_Age'] * datos['Dependent_count']

#Agregar una columna con la división de las columnas
datos2['Division'] = datos['Customer_Age'] / datos['Dependent_count']

# Mover la columna 'C' al inicio
columna = datos2.pop("Division")  # Elimina la columna y la guarda en una variable
datos2.insert(1, "Division", columna)  # Inserta la columna en la posición deseada
```

##### 6 Comandos de logica y agregar etiquetas a columa/variable

```python
#Eliminar observaciones si la columna 'Customer_Age' es mayor a 50
datos2 = datos[datos['Customer_Age'] < 50]

#Conservar observaciones si la columna 'Customer_Age' es mayor o igual a 50
datos2 = datos[datos['Customer_Age'] >= 50]

#Eliminar observaciones si la columna 'Customer_Age' es mayor a 50 y 
# la columna 'Dependent_count' es mayor a 2
datos2 = datos[(datos['Customer_Age'] < 50) & (datos['Dependent_count'] < 2)]

#Eliminar observaciones si la columna 'Customer_Age' es mayor a 50 o 
# la columna 'Dependent_count' es mayor a 2
datos2 = datos[(datos['Customer_Age'] < 50) | (datos['Dependent_count'] < 2)]

#Eliminar observaciones si la columna 'Dependent_count' es mayor a 2
datos2 = datos.drop(datos[datos['Dependent_count'] > 2].index)

#Eliminar observaciones si la columna 'Dependent_count' es mayor a 2 
# y la columna 'Education_Level' es igual a 'Graduate'
datos2 = datos.drop(datos[(datos['Dependent_count'] > 2) & (datos['Education_Level'] == 'Graduate')].index)

#Eliminar observaciones si la columna 'Dependent_count' 
# es mayor a 2 o la columna 'Education_Level' es igual a 'Graduate'
datos2 = datos.drop(datos[(datos['Dependent_count'] > 2) | (datos['Education_Level'] == 'Graduate')].index)

#Agregar una columna igual a 1 si la columna gender es igual a 'M' y 0 si es 'F'
datos2 = datos.drop(["Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1", "Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1"], axis = 1)
datos2 = datos.drop(["Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1", "Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2"], axis = 1)
datos2['Genero'] = datos2['Gender'].apply(lambda x: 1 if x == 'M' else 0)

#Etiquetar las observaciones de "Genero" como "Masculino" si la columna "Genero" es igual a 1 y "Femenino" si es igual a 0
datos2['Genero'] = datos2['Genero'].apply(lambda x: 'Masculino' if x == 1 else 'Femenino')
```

```python
# Eliminar columnas de la base de datos
datos2 = datos.drop(['CLIENTNUM', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], axis=1)

# Transformar la columna 'Education_Level' a categórica
datos2['Education_Level'] = datos2['Education_Level'].astype('category')

# Ver los valores que toma cada variable categórica
datos2['Education_Level'].unique()

# Reacomodar los valores de la variable categórica
datos2['Education_Level'] = datos2['Education_Level'].cat.set_categories(['Uneducated', 'High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate'], ordered=True)
datos2['Education_Level'].unique()
```
